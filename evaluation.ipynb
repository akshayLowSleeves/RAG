```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv() 
openai_api_key = os.getenv("OPENAI_API_KEY")
if not openai_api_key:
    raise ValueError("OPENAI_API_KEY environment variable not set")

client = OpenAI(api_key=openai_api_key)

def check_answer(question, answer, rag_answer):
    prompt = f"""
You are an answer checker. Given a question, its correct answer, and a predicted answer, determine if the predicted answer is factually correct compared to the original answer.

Instructions:
- If the predicted answer conveys the same main information as the correct answer (even if phrased differently), reply "correct".
- If the predicted answer is factually incorrect or significantly deviates from the correct answer, reply "wrong".
- Do not provide explanations or verbose responses. Only output one word: "correct" or "wrong".

Input format:
question: {question}
answer: {answer}
rag_answer: {rag_answer}

Output format:
[correct/wrong]
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini", 
        messages=[
            {"role": "system", "content": "You are an answer checker."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=1,  
        temperature=0  
    )

    result = response.choices[0].message.content.strip().lower()
    return result
```

```python
import time
result = {}

for k,v in loaded_data.items():
    result[k]=[]
    for x in v:

        question = x["question"]
        answer = x["answer"]
        rag_answer = x["rag_answer"]

        result_ = check_answer(question, answer, rag_answer)
        result[k].append(result_)
        print(k,result_) 
        time.sleep(1)
```

```python
result_acc = {}
for k,v in result.items():
    result_acc[k]={}
    length = len(v)
    corr = 0
    wrong = 0
    for x in v:
        if x=='correct':
            corr+=1
        if x=='wrong':
            wrong+=1
    result_acc[k]['total']=length
    result_acc[k]['corr']=corr
    result_acc[k]['wrong']=wrong
    result_acc[k]['acc'] = round((corr / length) * 100, 2)

result_acc 
```
